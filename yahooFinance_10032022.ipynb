```python
meta = yf.Ticker("META")
```

```python
meta.info
```

```python
hist = meta.history(period='1d', start='2018-01-01', end='2022-10-01')
```

```python
px = hist.Close
```

```python
hist.Close.plot()
```

```python
px.groupby(by=[px.index.year, px.index.month]).mean()
```
